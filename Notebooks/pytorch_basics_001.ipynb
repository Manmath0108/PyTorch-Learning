### Tensors Datatypes

NOTE --> Tensor Datatype is one of the 3 big errors one runs into with PyTorch and Deep Learning:




    1) Tensors of not the right Datatype
    2) Tensors not the right shape
    3) Tensors not on the right device

```python
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                                dtype=None,
                                device=None,
                                requires_grad=False)
float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
float_64_tensor = float_32_tensor.type(torch.float64)
float_64_tensor
```

```python
tensor_mult = float_16_tensor * float_32_tensor
tensor_mult
```

```python
tensor_mult_2 = float_32_tensor * float_64_tensor
tensor_mult_2
```

```python
tensor_mult.dtype
```

```python
tensor_a = torch.rand(3, 4)
print(tensor_a)
tensor_b = torch.rand(3, 4)
print(tensor_b)
result = tensor_a * tensor_b
print(result)
print(result.dtype)
```

### Reshaping, Stacking, Squeezing and Unsqueezing Tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - Retrun a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze - removes all `1` dimensions from a tensor
* Unsqueeze - add a `1` dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
import torch
x = torch.arange(1., 10.)
```

```python
# Reshape
x_reshaped = x.reshape(9, 1)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
#Stack
x_stacked_1 = torch.stack([x, x, x, x], dim=0)
x_stacked_1
```

```python
x_stacked_2 = torch.stack([x, x, x, x], dim=1)
x_stacked_2
```

```python
x_stacked_3 = torch.stack([x, x, x, x, x], dim=0)
x_stacked_3
```

```python
x = torch.zeros(2, 1, 2, 1, 2)
x, x.shape
```

```python
y = torch.squeeze(x)
y, y.shape
```

```python
y = torch.squeeze(x, 0)
y, y.size()
```

```python
y = torch.squeeze(x, 1)
y, y.size()
```
